```python
import os

# Path to dataset
data_path = "../data/leapGestRecog"

# List of folders
folders = os.listdir(data_path)
print(f"Folders: {folders}")
```

```python
# Count samples in each gesture folder
gesture_folders = []

for folder in folders:
    folder_path = os.path.join(data_path, folder)
    if os.path.isdir(folder_path):
        gestures = os.listdir(folder_path)
    for gesture in gestures:
            gesture_path = os.path.join(folder_path, gesture)
            if os.path.isdir(gesture_path):
                images = os.listdir(gesture_path)
                gesture_folders.append((folder, gesture, len(images)))


print("Sample counts:")
for s in gesture_folders[:5]:
    print(s)
```

```python
# Show sample images
import cv2
import matplotlib.pyplot as plt

# Example: show 1 random image
folder = folders[0]
gesture = os.listdir(os.path.join(data_path, folder))[0]
image_file = os.listdir(os.path.join(data_path, folder, gesture))[0]

image_path = os.path.join(data_path, folder, gesture, image_file)
print(f"Image path: {image_path}")

# Read image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Show image
plt.imshow(image, cmap='gray')
plt.title(f"Subject: {folder} | Gesture: {gesture}")
plt.show()

print(f"Image shape: {image.shape}")
```
